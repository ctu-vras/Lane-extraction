# Pytorch Autograd Framework


```python
import torch
import matplotlib.pyplot as plt
```

```python
x = torch.tensor(5.0, requires_grad=True)
y = torch.tensor(1.0)

optimizer = torch.optim.SGD([x], lr=0.01)
x_values = []
x_grad_values = []

for i in range(300):
    loss = (x - y)**2
    loss.backward()

    # store values before zeroing the gradient
    x_values.append(x.item())
    x_grad_values.append(x.grad.item())

    optimizer.step()
    optimizer.zero_grad()


```

```python
plt.plot(x_values, label='x')
plt.plot(x_grad_values, label='x_grad')
plt.title('x and x_grad values during optimization')
plt.grid()
plt.legend()
plt.show()
```
